```python
import pandas as pd
from Bio import SeqIO

df = pd.read_csv('/tscc/projects/ps-yeolab3/eboyle/encode/pipeline/adapters/katie/barcode_list.tsv', 
                   sep = '\t'
                  ).T.reset_index().T
```

```python
df = df[[0,2]]
```

```python
df
```

```python
from Bio.Seq import Seq
df['barcode']=df['seq'].str.replace('AGATCGGAAGAGCACACGTC/3AmMO/', '').str.replace('5Phos/NN','').apply(lambda seq: str(Seq(seq).reverse_complement()))
```
